```python
from google.colab import drive
drive.mount('/content/drive')

import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.dates as mdates
from scipy.stats import linregress

#Calling the data files
CO2graph = pd.read_csv('/content/drive/MyDrive/OceanAcidificationData/ESRLGlobalCO2.csv')
df = pd.read_csv('/content/drive/MyDrive/OceanAcidificationData/EPAGlobalCO2.csv')
pHAccepted = pd.read_csv('/content/drive/MyDrive/OceanAcidificationData/PhAccepted.csv')
pHAccepted['date'] = pd.to_datetime(pHAccepted['date'], format='%Y')
df['year'] = pd.to_datetime(df['year'], format='%Y')

#Setting some variables for future use
K1 = 1
K2 = 3 * (10 ** -3)
K3 = 2 * (10 ** -4)
K4 = 6 * (10 ** -11)

pHi = 8.11
Hi = 10 ** -8.11

#Creating several data frames that will be filled later
CO2df = pd.DataFrame()
pHdf = pd.DataFrame()
CO2values = pd.DataFrame()
CO2df['CO2 Injection'] = pd.Series(dtype='float64')
CO2df['Final H'] = pd.Series(dtype='float64')
pHdf['pH'] = pd.Series(dtype='float64')
CO2values['CO2 Values'] = pd.Series(dtype='float64')
pHdf.loc[0,'pH'] = pHi

#Graphing CO2 over time
plt.figure(figsize=(12, 9))
sns.lineplot(data=CO2graph, x='year', y='mean', label='Measured CO$_2$ Concentrations (ppm)', color='red')
plt.title('Worldwide Averaged CO$_2$ Concentrations between 1959 and 2022')
plt.legend(fontsize='small')
plt.ylabel('Carbon Dioxide Concentration (ppm)')
plt.xlabel('Date')
plt.grid(True)
plt.legend()
plt.savefig('Worldwide Averaged CO2 Concentrations between 1959 and 2022.png')
plt.show()

pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
print(CO2graph)

#Converting ppm values to molarity (mol/L) and saving those values to a data frame
for i in range(0,len(df)):
  ppm = df.loc[i,'mean']
  CO2i = ppm / 22.41 * (10 ** -6)
  CO2values.loc[i,'CO2 Values'] = CO2i

#Calculating values for carbonic acid conc., bicarbonate conc., and others, then adding those values to a data frame. The equilibrium H conc. was also calculated and added to a separate data frame
for j in range(0,len(df)):
  Carbonic = K2 * CO2values.loc[j,'CO2 Values']
  bicarb = K3 * Carbonic / (10 ** (-1 * pHdf.loc[j,'pH']))
  AddRowCO2 = {'date':df.loc[j,'year'], 'CO2 initial':df.loc[0,'mean'], 'Carbonic Acid':Carbonic, 'Bicarbonate':bicarb}
  CO2df = CO2df.append(AddRowCO2, ignore_index=True)
  if j != (len(df) - 1):
    CO2df.loc[j,'CO2 Injection'] = CO2values.loc[(j+1),'CO2 Values'] - CO2values.loc[j,'CO2 Values']
    H = 10 ** (pHdf.loc[j,'pH'] * -1)
  CO2df.loc[j,'Final H'] = H - (CO2df.loc[j,'Bicarbonate'] / 2) + math.sqrt((((CO2df.loc[j,'Bicarbonate']) ** 2) / 4) + (K1 * K2 * K3 * CO2df.loc[j,'CO2 Injection']) / (K1 + 1))
  pHdf.loc[j+1,'pH'] = math.log(CO2df.loc[j,'Final H'], 10) * -1

#Adding a date column to a data frame
pHdf['date'] = pd.Series(dtype='float64')
for i in range(0,len(df)):
  pHdf.loc[i,'date'] = df.loc[i,'year']

#Creating another data frame and adding in 3 empty columns to be filled later
CO2diff = pd.DataFrame()
CO2diff['Calculated CO2 Difference'] = pd.Series(dtype='float64')
CO2diff['Accepted CO2 Difference'] = pd.Series(dtype='float64')
CO2diff['Percentage of Acidification Due to CO2'] = pd.Series(dtype='float64')

#Calculating some values and adding them to a data frame then plotting it
for i in range(0,36):
  CO2diff.loc[i,'Calculated CO2 Difference'] = 8.11 - pHdf.loc[i,'pH']
  CO2diff.loc[i,'Accepted CO2 Difference'] = 8.11 - pHAccepted.loc[i,'ph']
  CO2diff.loc[i,'Percentage of Acidification Due to CO2'] = CO2diff.loc[i,'Calculated CO2 Difference'] / CO2diff.loc[i,'Accepted CO2 Difference']
AdjustedCO2diff = CO2diff[CO2diff['Percentage of Acidification Due to CO2'] != 0]
for i in range(0,len(df)):
  AdjustedCO2diff.loc[i,'date'] = df.loc[i,'year']


AdjustedCO2diff.set_index('date', inplace=True)
AdjustedCO2diff['numerical_date'] = mdates.date2num(AdjustedCO2diff.index)
# Set the size of the plot
plt.figure(figsize=(12, 9))
# Create the scatter plot
ax = sns.scatterplot(data=AdjustedCO2diff, x='numerical_date', y='Percentage of Acidification Due to CO2', label='Percent Ocean Acidification as a Product of Carbon Dioxide', color='darkblue')

# Add a line of best fit
mask = ~AdjustedCO2diff['Percentage of Acidification Due to CO2'].isna()
slope, intercept, r_value, p_value, std_err = linregress(AdjustedCO2diff['numerical_date'][mask], AdjustedCO2diff['Percentage of Acidification Due to CO2'][mask])
sns.regplot(data=AdjustedCO2diff, x='numerical_date', y='Percentage of Acidification Due to CO2', scatter=False, ax=ax)



tick_positions = range(0, len(AdjustedCO2diff.index), 3)
ax.set_xticks(AdjustedCO2diff['numerical_date'][tick_positions])
ax.set_xticklabels(AdjustedCO2diff.index[tick_positions].strftime('%Y'))

"""
if not (pd.isna(slope) or pd.isna(intercept)):
    equation = f'y = {slope:.5f}x + {intercept:.5f}\nR-squared = {r_value**2:.5f}'
    plt.text(0.05, 0.9, equation, transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
"""

if not (pd.isna(slope) or pd.isna(intercept)):
    equation = f'R-squared = {r_value**2:.5f}'
    plt.text(0.05, 0.9, equation, transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))




#plt.text(0.05, 0.9, f'R-squared = {r_value**2:.5f}', fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

plt.title('Relative Impact of Carbon Dioxide on Ocean Acidification')
plt.xlabel('Date')
plt.ylabel('$\Delta$pH$_{computed}$ / $\Delta$pH$_{actual}$')
plt.legend(labels=['Ratios','Linear Regression Line','Error'])
plt.savefig('Relative Impact of Carbon Dioxide on Ocean Acidification.png')
plt.show()

average_value = CO2diff['Percentage of Acidification Due to CO2'].mean()
print("Average:", average_value)

#Making another data frame
CO2adjusted = pd.DataFrame()
CO2adjusted['Adjusted CO2'] = pd.Series(dtype='float64')
CO2adjusted['date'] = pd.Series(dtype='float64')

#Adding a date column to a data frame
for i in range(0,len(df)):
  CO2adjusted.loc[i,'date'] = df.loc[i,'year']

#Adding the adjusted CO2 values to a data frame
for i in range(0,37):
  CO2adjusted.loc[i,'Adjusted CO2'] = 8.11 - ((8.11 - pHdf.loc[i,'pH']) / average_value)

#Graphing using seaborn
pHAccepted.set_index('date', inplace=True)
pHdf.set_index('date', inplace=True)
plt.figure(figsize=(12, 9))
sns.lineplot(data=pHdf, x=pHdf.index, y='pH', label='Surface Ocean pH Caused by CO$_2$')
sns.lineplot(data=pHAccepted, x=pHAccepted.index, y='ph', label='Accepted pH')
plt.title('Comparison Between Actual Surface Ocean pH\nand Ocean Acidification Caused by Carbon Dioxide')
plt.xlabel('Date')
plt.ylabel('pH')
plt.legend()
plt.savefig('Comparison Between Actual Surface Ocean pH and Ocean Acidification Caused by Carbon Dioxide.png')
plt.show()

CO2adjusted.set_index('date', inplace=True)
plt.figure(figsize=(12, 9))
sns.lineplot(data=CO2adjusted, x=CO2adjusted.index, y='Adjusted CO2', label='pH Estimate Based on CO$_2$ Concentrations\nand its Average Relative Impact on OA')
sns.lineplot(data=pHAccepted, x=pHAccepted.index, y='ph', label='Accepted pH')
plt.title('Comparison Between Actual Surface Ocean pH\nand an Estimate of Ocean pH Based on the Average Relative\nImpact of Carbon Dioxide on Ocean Acidification')
plt.xlabel('Date')
plt.ylabel('pH')
plt.legend()
plt.savefig('Comparison Between Actual Surface Ocean pH and an Estimate of Ocean pH Based on the Average Relative Impact of Carbon Dioxide on Ocean Acidification.png')
plt.show()


multidf = pd.DataFrame()
multidf['multipliers'] = pd.Series(dtype='float64')
AdjustedCO2diff.reset_index(drop=True, inplace=True)
pHdf.reset_index(drop=True, inplace=True)
pHdf.dropna(inplace=True)

for k in range(len(AdjustedCO2diff)):
  multiplier = AdjustedCO2diff.loc[k,'numerical_date'] * slope + intercept
  multidf.loc[k,'multipliers'] = 8.11 - ((8.11 - pHdf.loc[k,'pH']) / multiplier)

for i in range(0,len(df)):
  multidf.loc[i,'date'] = df.loc[i,'year']

multidf.loc[35,'multipliers'] = 8.0516155

multidf.set_index('date', inplace=True)
plt.figure(figsize=(12, 9))
sns.lineplot(data=multidf, x=multidf.index, y='multipliers', label='pH Estimate Based on CO$_2$ Concentrations\nand its Instantaneous Relative Impact on OA')
sns.lineplot(data=pHAccepted, x=pHAccepted.index, y='ph', label='Accepted pH')
plt.title('Comparison Between Actual Surface Ocean pH\nand an Estimate of Ocean pH Based on the Instantaneous Relative\nImpact of Carbon Dioxide on Ocean Acidification')
plt.xlabel('Date')
plt.ylabel('pH')
plt.legend()
plt.savefig('Comparison Between Actual Surface Ocean pH and an Estimate of Ocean pH Based on the Instantaneous Relative Impact of Carbon Dioxide on Ocean Acidification.png')
plt.show()

#####################################################################################################################################################################
#####################################################################################################################################################################
#####################################################################################################################################################################
#####################################################################################################################################################################

plt.style.use('fivethirtyeight')

# Load and preprocess the data
df = pd.read_csv('/content/drive/MyDrive/OceanAcidificationData/ESRLGlobalCO2.csv')
df_year_index = df.set_index('year')
df_year_index.index = pd.to_datetime(df_year_index.index, format='%Y')

df_year_index.drop(columns=['unc'], inplace=True)

color_pal = sns.color_palette()
#df_year_index.plot(figsize=(15, 5), color=color_pal[0], title='Carbon Dioxide Concentrations in ppm')

# Creating features and target variable
X = np.arange(len(df_year_index)).reshape(-1, 1)
y = df_year_index.values.ravel()

# Polynomial features transformation
degree = 2  # Change the degree as needed
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Extending the prediction to the year 2100
future_years = np.arange(len(df_year_index), 2261 - df_year_index.index[0].year).reshape(-1, 1)
future_years_poly = poly_features.transform(future_years)
predicted_co2 = model.predict(future_years_poly)

# Plotting the results including the extended forecast
plt.figure(figsize=(12, 9))
plt.scatter(df_year_index.index, y, label='Actual Data', color='red')
plt.plot(pd.date_range(start=df_year_index.index[0], periods=len(df_year_index) + len(future_years), freq='Y'), np.concatenate([y, predicted_co2]), color=sns.color_palette()[0], label='Second Degree Linear Regression Line')
plt.legend()
plt.xlabel('Date')
plt.ylabel('CO$_2$ Concentration (ppm)')
plt.title('Generating a Forecast for Future\nGaseous Carbon Dioxide Concentrations')
plt.savefig('Generating a Forecast for Future Gaseous Carbon Dioxide Concentrations.png')
plt.show()

print(predicted_co2)

# Creating a DataFrame for the polynomial regression forecast
forecast_years = pd.date_range(start=df_year_index.index[-1] + pd.Timedelta(days=365), periods=len(future_years), freq='Y')

forecast_df = pd.DataFrame()
forecast_df['year'] = forecast_years
forecast_df['prediction'] = predicted_co2

df['year'] = pd.to_datetime(df['year'], format='%Y')
df = df.rename(columns={'mean': 'CO2 Data'})
forecast_df = forecast_df.rename(columns={'prediction': 'CO2 Data'})

RealAndForecast = pd.concat([df, forecast_df])
after85 = RealAndForecast[RealAndForecast['year'] >= '1985-01-01']
after85 = after85.reset_index(drop=True)


##############################################################################################################################################################################################################


#Setting some variables for future use
K1 = 1
K2 = 3 * (10 ** -3)
K3 = 2 * (10 ** -4)
K4 = 6 * (10 ** -11)

pHi = 8.11
Hi = 10 ** -8.11

#Creating several data frames that will be filled later
CO2df2 = pd.DataFrame()
pHdf2 = pd.DataFrame()
CO2values2 = pd.DataFrame()
CO2df2['CO2 Injection'] = pd.Series(dtype='float64')
CO2df2['Final H'] = pd.Series(dtype='float64')
pHdf2['pH'] = pd.Series(dtype='float64')
CO2values2['CO2 Values'] = pd.Series(dtype='float64')
pHdf2.loc[0,'pH'] = pHi

#Converting ppm values to molarity (mol/L) and saving those values to a data frame
for i in range(0,len(after85)):
  ppm = after85.loc[i,'CO2 Data']
  CO2i2 = ppm / 22.41 * (10 ** -6)
  CO2values2.loc[i,'CO2 Values'] = CO2i2

#Calculating values for carbonic acid conc., bicarbonate conc., and others, then adding those values to a data frame. The equilibrium H conc. was also calculated and added to a separate data frame
for j in range(0,len(after85)):
  Carbonic2 = K2 * CO2values2.loc[j,'CO2 Values']
  bicarb2 = K3 * Carbonic2 / (10 ** (-1 * pHdf2.loc[j,'pH']))
  AddRowCO22 = {'date':after85.loc[j,'year'], 'CO2 initial':after85.loc[0,'CO2 Data'], 'Carbonic Acid':Carbonic2, 'Bicarbonate':bicarb2}
  CO2df2 = CO2df2.append(AddRowCO22, ignore_index=True)
  if j != (len(after85) - 1):
    CO2df2.loc[j,'CO2 Injection'] = CO2values2.loc[(j+1),'CO2 Values'] - CO2values2.loc[j,'CO2 Values']
    H2 = 10 ** (pHdf2.loc[j,'pH'] * -1)
  CO2df2.loc[j,'Final H'] = H2 - (CO2df2.loc[j,'Bicarbonate'] / 2) + math.sqrt((((CO2df2.loc[j,'Bicarbonate']) ** 2) / 4) + (K1 * K2 * K3 * CO2df2.loc[j,'CO2 Injection']) / (K1 + 1))
  pHdf2.loc[j+1,'pH'] = math.log(CO2df2.loc[j,'Final H'], 10) * -1

#Adding a date column to a data frame
pHdf2['date'] = pd.Series(dtype='float64')
for i in range(0,len(after85)):
  pHdf2.loc[i,'date'] = after85.loc[i,'year']



multiplier_df = pd.DataFrame()
multiplier_df['multipliers'] = pd.Series(dtype='float64')
multiplier_df['date'] = pd.Series(dtype='float64')
pHdf2.dropna(inplace=True)
pHdf2['numerical_date'] = mdates.date2num(pHdf2['date'])

for k in range(len(pHdf2)):
  multiplier = pHdf2.loc[k,'numerical_date'] * slope + intercept
  multiplier_df.loc[k,'multipliers'] = 8.11 - ((8.11 - pHdf2.loc[k,'pH']) / multiplier)

for i in range(0,len(pHdf2)):
  multiplier_df.loc[i,'date'] = pHdf2.loc[i,'date']


plt.figure(figsize=(12, 9))
multiplier_df.set_index('date', inplace=True)
sns.lineplot(data=multiplier_df, x=multiplier_df.index, y='multipliers', label='pH Estimate Based on the Instantaneous\nRelative Impact of CO$_2$ on OA')
plt.xlabel('Date')
plt.ylabel('pH')
plt.title('Constructing a pH Forecast According to the Instantaneous\nRelative Impact of Carbon Dioxide on Ocean Acidification')
plt.legend()
plt.savefig('Constructing a pH Forecast According to the Instantaneous Relative Impact of Carbon Dioxide on Ocean Acidification')
plt.show()

pd.set_option('display.max_rows', None)
print(multiplier_df)
```
